# 异步电机转矩
$T=\frac{m1p}{\omega_{1}}U_{1}^{2}\frac{\frac{r_{2}^{'}}{s}} {(r_{1}+c1\frac{r_{2}^{'}}{s})^{2}+(x_{1}+c_{1}x_{2}^{'})^{2}}$



```python
from ipywidgets import interact, interactive
from IPython.display import clear_output, display, HTML
m1=3
p=6
f=50
U1=3000
r1=0.42
x1=2
r2=0.45
x2=2
xm=48.7

omega1= 6.28*f
s=(0.05)


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def motor(r2):
            c1=1+x1/xm
            T=lambda s:(m1*p/omega1)*U1**2*(r2/s)/((r1+(c1*r2/s))**2+(x1+c1*x2)**2)
            s = np.linspace(-2, 2.1 ,1000)
            y=T(s)
            plt.figure(figsize=(20,10))

            plt.xlabel("$s$")
            plt.ylabel("T")
            plt.title("T-s")
            plt.grid(b='on')

            ax=plt.axes()
            ax.spines['right'].set_color('none')
            ax.spines['top'].set_color('none')
            ax.xaxis.set_ticks_position('bottom')
            ax.spines['bottom'].set_position(('data',0))
            ax.yaxis.set_ticks_position('left')
            ax.spines['left'].set_position(('data',0))
            ax.axes.invert_xaxis()
            ysk=c1*r2/np.sqrt(r1**2+(x1+c1*x2)**2)
            ysk = np.linspace(ysk,ysk,1000)
            plt.plot(s,T(ysk),label="$T_{max}$",color="black",linewidth=2)
           
            y=T(s)
            line2=plt.plot(s,y,label="$T_{2}(s)$",color="blue",linewidth=2)
            plt.legend()
            plt.show()
            
            
w = interactive(motor, r2=(0.45,4))
display(w)
```
